This function is to plot the confusion matrix.


```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

## TASK  1

## TASK  2


Standardize the data in <code>X</code> then reassign it to the variable  <code>X</code> using the transform provided below.


```python
# students get this 
transform = preprocessing.StandardScaler()
# Estandarizamos X y reasignamos
X = transform.fit_transform(X)
```

## TASK  3


Use the function train_test_split to split the data X and Y into training and test data. Set the parameter test_size to  0.2 and random_state to 2. The training data and test data should be assigned to the following labels.


<code>X_train, X_test, Y_train, Y_test</code>


```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=2
)
```

## TASK  8


Create a decision tree classifier object then  create a  <code>GridSearchCV</code> object  <code>tree_cv</code> with cv = 10.  Fit the object to find the best parameters from the dictionary <code>parameters</code>.


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

parameters = {
    'criterion': ['gini', 'entropy'],         # 'entropy' sigue siendo válido
    'splitter': ['best', 'random'],
    'max_depth': [2*n for n in range(1, 10)], # 2,4,6,8,10,12,14,16,18
    'max_features': [None, 'sqrt', 'log2'],   # <-- quitar 'auto'
    'min_samples_leaf': [1, 2, 4],
    'min_samples_split': [2, 5, 10]
}
```

```python
tree = DecisionTreeClassifier(random_state=2)
tree_cv = GridSearchCV(tree, parameters, cv=10, n_jobs=-1)  # opcional: error_score='raise'
tree_cv.fit(X_train, Y_train)
```

```python
print("tuned hyperparameters :(best parameters) ", tree_cv.best_params_)
print("accuracy :", tree_cv.best_score_)
```

## TASK  9


Calculate the accuracy of tree_cv on the test data using the method <code>score</code>:


```python
test_acc = tree_cv.score(X_test, Y_test)
print("Test set accuracy:", test_acc)
```

We can plot the confusion matrix


```python
yhat = tree_cv.predict(X_test)
plot_confusion_matrix(Y_test, yhat)
```

## TASK  10


Create a k nearest neighbors object then  create a  <code>GridSearchCV</code> object  <code>knn_cv</code> with cv = 10.  Fit the object to find the best parameters from the dictionary <code>parameters</code>.


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Búsqueda de hiperparámetros
parameters = {
    'n_neighbors': list(range(1, 11)),          # 1 a 10
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'p': [1, 2]                                  # p=1 Manhattan, p=2 Euclidiana
}

knn = KNeighborsClassifier()

knn_cv = GridSearchCV(knn, parameters, cv=10, n_jobs=-1)
knn_cv.fit(X_train, Y_train)

print("tuned hyperparameters :(best parameters) ", knn_cv.best_params_)
print("accuracy :", knn_cv.best_score_)
```

## TASK  11


Calculate the accuracy of knn_cv on the test data using the method <code>score</code>:


```python
# Accuracy en el conjunto de test
test_acc = knn_cv.score(X_test, Y_test)
print("Test set accuracy:", test_acc)

# Matriz de confusión
yhat = knn_cv.predict(X_test)
plot_confusion_matrix(Y_test, yhat)
```

We can plot the confusion matrix


## TASK  12


Find the method performs best:


```python
# --- Task 12: ¿qué método rinde mejor en test? ---

scores_test = {
    "Logistic Regression": logreg_cv.score(X_test, Y_test),
    "SVM":                 svm_cv.score(X_test, Y_test),
    "Decision Tree":       tree_cv.score(X_test, Y_test),
    "KNN":                 knn_cv.score(X_test, Y_test),
}

for name, sc in scores_test.items():
    print(f"{name:>16}: test accuracy = {sc:.4f}")

best_model_name = max(scores_test, key=scores_test.get)
print("\n>> Best method on the test set:", best_model_name,
      f"({scores_test[best_model_name]:.4f})")
```

```python
import matplotlib.pyplot as plt

scores = {
    "LogReg": logreg_cv.best_score_,
    "SVM":    svm_cv.best_score_,
    "Tree":   tree_cv.best_score_,
    "KNN":    knn_cv.best_score_,
}

# (optional) sort by accuracy
order = sorted(scores, key=scores.get, reverse=True)
names = order
vals  = [scores[k] for k in order]

plt.figure(figsize=(7,4))
bars = plt.bar(names, vals)
plt.ylim(0, 1)
plt.ylabel("Accuracy")
plt.title("Cross-validated accuracy by model")

# annotate bars
for b, v in zip(bars, vals):
    plt.text(b.get_x()+b.get_width()/2, v+0.005, f"{v:.3f}", ha="center", va="bottom")

plt.show()
```
